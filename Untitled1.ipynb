```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# NOTE: This code assumes the DataFrame 'df' was loaded in Question 5.

# 1. Encode the target variable 'PriceRange' (Necessary for scikit-learn)
le = LabelEncoder()
df['PriceRange_encoded'] = le.fit_transform(df['PriceRange'])
# Store the original class names for clear reports/plots
price_classes = le.classes_

# 2. Define Features (X) and Target (y) from Q3
X = df[['RAM', 'BatteryPower']].values
y = df['PriceRange_encoded'].values

# 3. Split data into training (70%) and testing (30%) sets (Essential for proper evaluation)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 4. Retrain the Linear SVM Model (Consistent with Q4's likely configuration)
svm_model_q6 = SVC(kernel="linear", C=1, random_state=42)
svm_model_q6.fit(X_train, y_train)

# 5. Make Predictions on the Test Set
y_pred = svm_model_q6.predict(X_test)

# 6. Print Evaluation Metrics
print("## Model Evaluation for Linear SVM (Features: RAM, BatteryPower)")
print("-" * 60)

# Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"Overall Accuracy Score on Test Data: {accuracy:.4f}\n")

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred,
                             target_names=price_classes,
                             zero_division=0))

# 7. Plot Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=price_classes, yticklabels=price_classes)
plt.title('Confusion Matrix: Linear SVM Performance')
plt.xlabel('Predicted Price Range')
plt.ylabel('Actual Price Range')
plt.show()
```
